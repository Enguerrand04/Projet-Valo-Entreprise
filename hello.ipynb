# Simple Hello World

```java
import java.util.InputMismatchException;
import java.util.Scanner;

public class Main {
    public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);
        int vdAm = 0;
        int rexAm = 0;
        boolean saisieValide = false;

        // Saisie pour vd am
        while (!saisieValide) {
            try {
                System.out.print("Veuillez saisir la valeur numérique de 'vd am': ");
                vdAm = scanner.nextInt();
                saisieValide = true;
            } catch (InputMismatchException e) {
                System.out.println("Erreur : Veuillez entrer un nombre entier valide.");
                scanner.next(); // Vide le buffer
            }
        }

        saisieValide = false; // Réinitialise pour la deuxième saisie

        // Saisie pour REX am
        while (!saisieValide) {
            try {
                System.out.print("Veuillez saisir la valeur numérique de 'REX am': ");
                rexAm = scanner.nextInt();
                saisieValide = true;
            } catch (InputMismatchException e) {
                System.out.println("Erreur : Veuillez entrer un nombre entier valide.");
                scanner.next(); // Vide le buffer
            }
        }

        System.out.println("La valeur de 'vd am' est : " + vdAm);
        System.out.println("La valeur de 'REX am' est : " + rexAm);

        scanner.close();
    }
}
```
